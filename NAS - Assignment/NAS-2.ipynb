PERFORMANCE OF BOYS AND GIRLS ACROSS STATES

Importing the necessary libraries and examining the dataset

```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
marks = pd.read_csv('nas-pupil-marks.csv')
list(marks)
```

```python
df=marks.rename(columns={'Maths %':'Maths'})
df.describe()
```

Finding the total number of null vaules in the dataset

```python
df.isnull().sum()
```

Removing the null values from dataset and replacing them by their mean

```python
df['Use computer'] = df['Use computer'].map({"Yes":1,"No":0})
null_col = ['Maths', 'Reading %', 'Science %', 'Social %', 'Use computer']

for col in null_col:
    df[col].fillna((df[col].mean()), inplace=True)
```

```python
state=['AN','AP','AR','BR','CG','CH','DD','DL','DN','GA','GJ','HP','HR','JH','JK','KA','KL','MG','MH','MN','MP','MZ','NG','OR','PB','PY','RJ','SK','TN','TR','UK','UP','WB']

num_state = len(state)
print num_state
```

```python
boys_cnt=np.zeros(num_state)
boys_num=np.zeros(num_state)
girls_cnt=np.zeros(num_state)
girls_num=np.zeros(num_state)
```

```python
boys_cnt_sci=np.zeros(num_state)
boys_num_sci=np.zeros(num_state)
girls_cnt_sci=np.zeros(num_state)
girls_num_sci=np.zeros(num_state)

boys_cnt_soc=np.zeros(num_state)
boys_num_soc=np.zeros(num_state)
girls_cnt_soc=np.zeros(num_state)
girls_num_soc=np.zeros(num_state)
```

Next, computing the average score in mathematics of boys and girls across various states of India, also maintaing a count of states where boys perform better than girls and vice versa

Finding out the average Score obtained by Boys and Girls in Maths, Science and Social Science.

```python
for i in range(0, len(marks)):
    if (math.isnan(marks.iloc[i]['Maths %']))==False:
            val=state.index(marks.iloc[i]['State'])#=='AN':
            if marks.iloc[i]['Gender']==1:
                boys_cnt[val]=boys_cnt[val]+1
                boys_num[val]=boys_num[val]+marks.iloc[i]['Maths %']
            if marks.iloc[i]['Gender']==2:
                girls_cnt[val]=girls_cnt[val]+1
                girls_num[val]=girls_num[val]+marks.iloc[i]['Maths %']

print "here"
boys_outperform=0   #number of outperforming boys
girls_outperform=0 #number of outperforming girls
maxdiff=0.00 #maximum difference between the girls and boys df across states
```

```python
for i in range(0, len(marks)):
    if (math.isnan(marks.iloc[i]['Science %']))==False:
            val=state.index(marks.iloc[i]['State'])#=='AN':
            if marks.iloc[i]['Gender']==1:
                boys_cnt_sci[val]=boys_cnt_sci[val]+1
                boys_num_sci[val]=boys_num_sci[val]+marks.iloc[i]['Science %']
            if marks.iloc[i]['Gender']==2:
                girls_cnt_sci[val]=girls_cnt_sci[val]+1
                girls_num_sci[val]=girls_num_sci[val]+marks.iloc[i]['Science %']
                
    elif (math.isnan(marks.iloc[i]['Social %']))==False:
            val=state.index(marks.iloc[i]['State'])#=='AN':
            if marks.iloc[i]['Gender']==1:
                boys_cnt_soc[val]=boys_cnt_soc[val]+1
                boys_num_soc[val]=boys_num_soc[val]+marks.iloc[i]['Social %']
            if marks.iloc[i]['Gender']==2:
                girls_cnt_soc[val]=girls_cnt_soc[val]+1
                girls_num_soc[val]=girls_num_soc[val]+marks.iloc[i]['Social %']


boys_outperform_sci=0   #number of outperforming boys in science
girls_outperform_sci=0 #number of outperforming girls in science
boys_outperform_soc=0   #number of outperforming boys in social science
girls_outperform_soc=0 #number of outperforming girls in social science
```

Plotting the performance of Boys vs Girls for each state of India.

```python
st=''
for i in range(0,len(state)):
    print 'Avg of boys in ',state[i],'is - ',((boys_num[i]/boys_cnt[i]))
    print 'Avg of girls in ',state[i],'is - ',((girls_num[i]/girls_cnt[i]))
    boys = (boys_num[i]/boys_cnt[i])
    girls = (girls_num[i]/girls_cnt[i])
    
    labels = ('Boys Performance', 'Girls Performance')
    ypos = np.arange(len(labels))
    perf = [boys, girls]
    plt.bar(ypos, perf, align='center', color=['blue', 'pink'])
    plt.xticks(ypos, labels)
    plt.ylabel('Avg Performance')
    plt.show()
    v1=(boys_num[i]/boys_cnt[i])
    v2=(girls_num[i]/girls_cnt[i])
    if abs(v2-v1)>maxdiff:
        maxdiff=abs(v2-v1)
        st=state[i]
    if  v1> v2:
        boys_outperform=boys_outperform+1
    else :
        girls_outperform=girls_outperform+1
    print '\n'

print 'Number of states where boys outperform ',boys_outperform
print 'Number of states where girls outperform ',girls_outperform    
print 'The maximum difference between the performance of girls and boys is - ',maxdiff,' which is in the state ', st
print st
```
